# 股票的研究-作为暑假作业

## 投资

### 现价

投资是将现期的收入放到未来以获取长期来看更高的收入流的方法。一个首要问题是从现在的效用来看，未来是否与现在等价。如假设每天都有99%的概率死亡，那么效用最大化的策略几乎就是把到手的现金全部消费掉。所以未来的价值似乎要折价计算。不过从另一个角度看，即使的确假设等价，未来的价值往往也要折价，这是因为一个资金如果不投入或投入到并非最优的资产上，那么它就比最优资产在未来获得的总收益要低。所以我们往往根据某个折现率对未来的资产进行折现。这是因为

$$
P_0 k^n = \sum_{i=1}^n P_i k^{n-i}
$$

上式无非是要求一个现价$P_0$应该在按照一个增长率$k$指数增长n论后，与每轮资产会提供的收益$P_i$再按$k$增长剩余轮数后得到的收益应该相等（这里忽略了资产的残余价值）。可以把此处的资产看成机器，机器每轮产生收益P_i，$P_0$是购入价格，$k$则是随手可以达到的利率/收益，如定期存款利率。这是个有意义的约束，因为实际现价低于理论现价，则可以买入机器，之后产生更高的累计收益。若实际现价高于理论现价（或者说n期累计收益），如果你有机器，可以卖出机器，一般来说，在金融发达的市场，你甚至可以卖空来获取卖出机器的额外收益，这就是套利。在这个极其简化的模型中，我们可以看出若当理论现价与实际现价不一致时，就产生了套利机会。

### 风险与收益

上述模型在各个方面都过于简化了，比如它认为有一个固定的$k$，比如它是离散的，可以想象变分法就是用来解决这种问题的，同样，按照现实的复杂度，变分法实际上没法用。这是十分可笑的情况，对于越复杂的情况，我们反而被迫只能使用越简单的方法，就像在生物和社会科学中发生的那样。

真正致命的一点也许是它的随机性，也就是我们对未来信息的本质无知这一点没有表现出来。连续可以用离散模拟，固定参数可以分段分治，但随机性的缺乏却造成了致命的打击。直观上，我们认为一个稳定收益为10，与一个期望是10，但方差是400的投资相比更佳。这可以作不同的解释，如效用边际收益递减，这就使投资经效用映射后的随机变量，后者的期望要低于前者。不过很难说效用曲线的形态到底如何，总之我们这里指出对于投资来说，一般人都是风险厌恶者。当然也可以注意到很多人购买彩票或进行甚至收益本身就是负的赌博，这也许是因为收益分布的华丽的尾部为他们提供了那么一小点正效用，不过说到底他们也不会将其作为一种大规模的投资手段。

既然我们固有的希望降低随机变量的风险而喜欢提高其期望，其结果就是我们在两者间寻求一个trade off，可以使用夏普比率来刻画一项投资的额外吸引力：

$$
sharpen(I) = \frac{E(I)-r_f}{\sigma_I}
$$

即作为随机变量的某项投资$I$的期望减无风险利率$r_f$与其标准差之比。虽然这项汇总似乎过于简单，但它起码做到了有正确的量纲与两个正确的单调性，如果我们把大的夏普比率看做有吸引力的话。

此时，我们就可以比较两个投资的夏普比率，来决定哪个相对于它的风险有着更高的收益。对于收益本身小的那一方，某种意义上我们相信它可以保持相同的夏普比率“平移”到另一个投资相同风险的水平，此时前者的收益率就会高于后者。

在建模为正态分布的投资收益，夏普比率的平移可以表示为直接增加一项资产的投资数量（假如没有发生边际效应递减），从总量而非单位的角度看。就会得到如此关系:

$$
I_1 \sim N(E_1,\sigma_1^2) \\
I_2 \sim N(E_2,\sigma_2^2) \\
E_1 < E_2 \\
\frac{E_1-r_f}{\sigma_1} > \frac{E_2-r_f}{\sigma_2} \\
$$

令

$$
k=\frac{\sigma_2}{\sigma_1}
$$

则有

$$
I_3 = kI_1 \sim N(kE_1, k^2\sigma_1^2) = N(kE_1, \sigma_2^2)
$$



只需证明$kE_1 > E_2$,只需证明,又有题设知

$$
k(E_1 - r_f) > E_2 - r_f \\
\sigma_1 < \sigma_2 \\ 
k > 1 \\
再假设 r_f >0 \\
kE_1 - r_f > kE_1 - kr_f = k(E_1 - r_f) > E_2 - r_f
$$

这就证明了原命题。

## 股票与投资组合

当企业主发现自己的企业有着较高的夏普比率或等效风险的收益率时。他会倾向于把更多利润投入到再生产中，但是否他就对其他夏普比率更低的投资机会视而不见呢？事实也并非如此。事实上，假设资产收益又建模为正太分布，投资于同一项资产将使方差以平方速度增加。

$$
I \sim N(E,\sigma^2) \\
kI \sim N(kE,k^2\sigma^2)
$$

而购买收益独立的等参数资产则只使方差线性增加

$$
\sum_{i=1}^n I \sim N(kE,k\sigma^2)
$$

从平均风险来看，投资同一项资产平均风险不变，而投资不同的收益独立的资产则使平均风险以$\frac{\sigma^2}{n}$速度下降。当然事实上很少有完全独立甚至负相关收益的资产，但只要它们的相关系数不为+1.就可以在等效风险上取得更高收益，或者使投资组合达到更高的夏普比率。

在有多项资产的情况下，将它们建模为多元正态分布
$$
\bar{I} \sim N(E,\Sigma)
$$

我们可以求出其中夏普比率最高的投资组合。该组合与夏普比率不定的无风险利率线性插值形成的资本市场线(CML)上拥有相同的（最高的）夏普比率。投资者会根据它的风险厌恶程度从中选出一个点作为最终的投资组合。

从上面的论述我们可以看出，某种意义上可选的投资越多越好，因为相比之下可选的少是相对于可选多下以更多约束进行最优化，结果不可能更好。出于这种考虑，企业主之间互相投资就成了有吸引力的事情。但一个企业向其他企业可以提供不同类型的资产进行投资：

1.债券，公司可以提供利率为$r_p$的债券，只要利率高于无风险利率$r_f$就是“有意义”的，它的现值会受其风险影响，越大自然价格越低。

2.权益，基本上就是股票，权益不保证派发类似债券利息的现金股利，也不存在最后要归还的纸面价值（虽然从折旧上看，如果债券很长期，最后归还的纸面价值也是现价中不重要的构成部分）。它是公司资产的索要权和投票权（不考虑优先股），股票有类似资产现价的现值估价模型(股利折现模型和自由现金流折现模型)，以此我们就可以适用之前讨论的套利之类的概念于股票上。

## 基本面分析与财务分析

股票的内在价值，如果是建立在未来的股利发放上，我们可以给出股利折现模型：

$$
P_0 = \sum_{i=1}^n \frac{I_i}{(1+k)^i} + \frac{I_{n+1} + P_{n+1}}{(1+k)^{n+1}}
$$

其中$P_0$是期初内在价值，$I_i$为i期股利，$P_{n+1}$为期末残值。折现率$k$称为市场资本化率，我们先不管这个量是怎么来的。

这个公式看上去不错，只是如果我们要使用它的话，就必须估计未来若干年的股利和期末残值，这和直接估计$P_0$好像也没有简化多少。

于是加上两个关键假设，持续经营和利率以一个恒定速度$g$增长（当然也可以有$g=0$），这时我们可以将残值无限展开，利用等比级数得到其值。于是内在价值的估计变为只需要三个参数。

$$
P_0 = \frac{I_1}{k-g}
$$

但是毕竟在折现下最近几年是最重要的，而且我们也有更多信息去表示此时的变异。这时就产生一些多阶段模型，可能前几年是给定值，有段时间是插值，最后时间又变为恒定速度增长模型了。

有些公司几乎从来不发放股利，这时可以使用自由现金流折现模型来评估内在价值。它与股利折现模型几乎完全一样，只是折现的对象变成了自由现金流，某种意义上那个就是可用来发股利也可不发的那个钱，只不过某些期望长期增长的公司选择不发而已，而折现率也从市场资本化率$k$改为了加权资本成本$WACC$，主要就是在$k$的基础上考虑了债务比例和债务利息的影响。

自由现金流比较有利于所谓财务模型给出预测值，财务模型通常是一个看上去很多参数，实际就是人工调参的拍脑袋的活，分析师往往根据所分析公司自己的说法凑参数形成研报，无论如何，用这些迷之参数可以用来对上述模型进行估计，从而得到一个所谓基本面分析所分析出的内在价值/现值。由于公司自己的说法通常围绕着现价，所以分析师建模的结果也的确通常围绕着观测现价，给人一种基本面分析也不是太不靠谱的感觉。



## 资本资产定价模型

由纯无风险利率策略到有效边界上最高夏普比率两点线性插值形成的资本市场线CML，其可以通过将各股/证券建模为多元正态分布建模得到。由于显然的复杂性原因，实际上是不可能的，甚至我们会先假设一个资本市场线CML，去反而估计那个理论多元整体分布的某些参数（通常是收益率，即期望向量）。这是我们假设指数，什么上证指数之类的表示了最高夏普比率的点，与3个月定期存款利率的月化版本作为无风险利率，得到估计的资本市场线CML。然后再根据理论去找出参数的值。

### 从资本市场线CML到证券市场线SML

资本市场线CML是无风险投资与最优夏普比率策略（下面称为市场组合）线性插值得到的，这个线满足夏普比率都是一致最优的。我们以类似方法构成一组证券市场线SML，是由投资某个特定股票的策略与最优夏普比率策略插值得到的。不过虽然策略是插值得到的，但它并不是直线，夏普比率也不是恒定的。事实上：

设$r_i$是个股$i$的收益率/利率。$r_m$是市场组合利率，$p$为投资于个股$i$的比例。一个给定$i$后，由$p$决定的策略$c$的收益为

$$
r_c = p r_i + (1-p) r_m
$$

数量特征

$$
E(r_c) = p E(r_i) + (1-p) E(r_m) \\
var(c) = p^2 var(r_i) + (1-p)^2 var(r_m) + 2p(1-p)cov(r_i,r_m) \\
std(c) = \sqrt{var(c)}
$$

给定$p$点的SML的斜率是

$$
\frac{d}{dp} E(r_c) = E(r_i) -E(r_m) \\
\frac{d}{dp} std(c) = \frac{1}{2} \frac{1}{\sqrt{var(c)}} \left( 2p var(r_i) - 2(1-p) var(r_m) + (2-4p) cov(r_i,r_m) \right) \\
\frac{dE(r_c)}{dstd(c)} = \frac{\frac{dE(r_c)}{dp} }{\frac{dstd(c)}{dp} } = \frac{2(E(r_i) - E(r_m))\sqrt{var(c)}}{ 2p var(r_i) - 2(1-p) var(r_m) + (2-4p) cov(r_i,r_m)}
$$



给出一个最关键的约束，它在最优点上重合且相切。故

$$
\frac{dE(r_c)}{dstd(c)}\bigg|_{c = 0} = \frac{2(E(r_i) - E(r_m)) \sqrt{var(r_m)}}{-2var(r_m)+2cov(r_i,r_m)} =
\frac{(E(r_i)-E(r_m))std(r_m)}{cov(r_i,r_m) - var(r_m)} =
\frac{E(r_m)-r_f}{std(r_m)}
$$

等价变换得
$$
E(r_i) - r_f = \frac{cov(r_i,r_m)}{var(r_m)} \left( E(r_m) - r_f \right)
$$

此式所展现的个别资产相对于无风险利率$r_f$的风险溢价与常数$\frac{cov(r_i,r_m)}{var(r_m)}$与市场组合自己的风险溢价的关系就是所谓的（个别证券$i$）证券市场线SML。

### beta系数与alpha

常数$\frac{cov(r_i,r_m)}{var(r_m)}$称为证券/个股的beta系数。考虑个股期望收益率$E(r-i) - r_f$与市场组合期望收益率$E(r_m) - r_f$的一元回归模型。

$$
E(r-i) - r_f = \alpha + \beta(E(r_m) - r_f)
$$

甚至$beta$系数的表达式都与一元回归模型的是一样的。

可以发现，资本资产定价模型实际预测了$\alpha = 0$。该模型还说明了个股超额收益完全由市场组合的超额收益以一个系数$\beta$提供或减少。没有什么“内部增长率”。而当我们的确用带$\alpha$的模型去回归这些股票$i$时，我们虽然很多时候$\alpha$的确显著不为0，符合理论预期。但也经常得到显著不为0的情况，此时称这个$\alpha$为超额收益率，资本资产定价模型无非断言超额收益率为0.

各个资产的收益通过两个方式与市场组合发生关系。一是各个资产收益和相关结构给定后，生成的市场组合。二是通过上面理论推导，发现单项资产的期望收益率与市场组合收益率的联系。利用后者就可以在实际上不知道前者（当然也就不知道市场组合是如何产生的）的情况下，直接估计市场组合，然后估计市场组合收益率与单项组合收益率的关系（beta值），这个beta值一定程度上反映的是之前忽视的相关结构的信息。

可以考虑一个资产自己发生变化（我们只是知道这个资产本身发生了变化，不知道相关结构之类的信息）。看看资本资产定价模型是如何在这种情况下变化的。
首先，如果一个资产的期望收益率上升，则

1.市场组合会包含更多的该项资产。
2.市场组合中该项资产内部的期望收益上升。

这导致

1.资产与市场组合的协方差上升，使得beta系数上升。
2.资产的超额收益率本身上升。

于是对于特定的SML

$$
Er_i - r_f = \beta(Er_m - r_f)
$$

我们看到了左边的增大如何导致右边两个方向增大而适应了变化。

那么一项资产本身的方差变大了呢？这当然与引发了协方差矩阵的变化，我们并不具体知道协方差矩阵的形态，我们只知道它发生了某种变化，其中的一个特征是对角线上的一个元素变大了。我们也可以保持相关矩阵不变，再代入方差列去生成协方差矩阵，此时我们设相关矩阵本身不变而方差列的某个元增大了。

显然，方程两边都不应该有任何变化，但beta内部结构也许发生了变化。显然，如果持有比还是一样的，则方差会变大，但协方差也会变大。如果持有比因此变小，两方面效应可能使变大变小不能确定，但是总之beta值不变。

当所有人都按照真实模型$N(E,\Sigma)$分析出并购买相同的市场组合时，他们在真实模型里的期望收益率，每个人对其期望收益率都是一样的。单个资产的夏普比率高的并不会因为它高就造成价格上升而期望收益下降而使得夏普比率下降。而是因为它的高在市场组合中有一个比较高的比例（与高的beta值，同时把超额收益往自己那拉），同时夏普比例低的只是比例较低而非完全不购买，因为我们要利用其中的独立性因素来降低风险。我们对这样一个理想的系统观测若干轮，进行回归，会很大概率发现beta值与真实模型中的非常接近，而alpha值也的确不显著偏离0.

这的确是一十分理想的情景，然而它在现实中的第一个失败点就是回归的人并不知道真正的市场组合是什么（如果有），从而使用指数之类的也许有系统偏差的代理变量去回归，既然是有系统偏差，那出什么错误都有可能。

第二，如果投资人自己也不知道真实模型是什么，乃至每个人估计出了不一样的模型，那么即使他们还是按照最大化夏普比率方法去投资市场组合。他们也会对同一项资产估计出不同的beta值，这会带来不同的现值的估计（假如之后的收益容易计算的话）。这可能会带来价格的变化，从而使真实模型发生变化，于是我们可能在一个参数变动的数据中试图估计出不变的参数，那当然不会得到靠谱的结果。

第三，即使投资人知道真实模型，但出于某种原因不进行夏普比例最大化的决策（比如非线性的风险厌恶等），那整个理论自然完全失效，关系根本无从找起。

第四，真实模型要是自己的参数在变化，直接使用假设参数不变的回归显然是不合理的。对此做一些只用三年五年的限制显然并没有什么本质理由。

### FAMA三因素模型

资本资产定价模型可以看成两种东西，一个是在极其严格的假设下的真实模型，另一个是只包含了市场组合/指数的一个回归模型。后者中使用市场组合，指数或宏观经济变量并没有什么本质区别，回归模型无非是寻求相关性或做一些贫乏的预测而已，其中市场组合作为变量无非是讲了一个好故事，在那个幻想世界中，市场组合的超额收益率真的可以完全解释个股的超额收益率。

既然资本资产定价模型已经退化成了讲故事的回归模型，我们就可以试图加入任何被我们认为有助于解释股票收益的变量来扩展它。FAMA发现各种其他因素中，公司规模与账面-市场比例（B/M）.其与市场组合超额收益率分别构成了系统风险溢价，规模溢价与市场价值比率溢价（价值溢价）。当它们对应的beta值显著时，称为这一因素存在风险敞口等。

值得注意的是，一般的回归模型的数据矩阵的每一行数据通常是个体“自己的”参数，如年龄，性别等，估计的参数则构成所有这些个体参数如何决定个体的另一个属性的结构。照搬到多因素模型上看上去应该拿企业自己的市值，B/M比例来当数据矩阵的行一样，然而却并非如此，首先，估计参数是每个股票都有自己的一份的，不在各个股票间共享，一定说共享的话，共享的是一个股票自己的时间序列。而数据矩阵的一行全部由外部/全局因素构成，和公司个体没有任何关系，我们估计的参数是用来解释总溢价如何与外部/全局/系统因素的溢价所确定的，这与企业自身而非外部/全局/系统的某个投资组合的溢价没有关系，溢价差异的来源（比如企业自己的B/M比例本身），都是不重要的，关键是如何响应外部溢价，这是个回避复杂度的方法。在此方法下，我们假设一个股票的反应程度在所回归的时间序列长度上不变，变得是外部溢价本身的值，至于企业自己这些因素是什么，无关紧要。

那么我们就必须估计外部规模溢价与B/M因素的风险溢价如何表示。我们可能想到，外部规模溢价使得越小的企业收益率越高（因为风险大），这个差别就可以看成规模溢价，可是如何刻画这一点呢？一个十分主观的方法是人工划分出大企业和小企业，分别以其构建一个投资组合（比如简单平均），小的收益率减去大的收益率就可以搞出这个指数。对于B/M比例，反映的是公司的“价值性”，它与E/P,C/P这些指标类似，其值越大表明公司越倾向于价值型而非增长型（E/P大表明投资者对公司自己评估的资产量（资产负债表上的账面价值）的估价越低。）。这个低某种意义上可能就反映了基本面的对价格的低估，之后可能获取更高的资本利得。也可能估价低反映了对企业长期发展不看好，这个价格相当于只在买股利来抵偿可能更高资本利得的损失，把企业当现金牛。

于是我们得到了回归方程

$$
r_i - r_f = \alpha_i + \beta_i(r_m - r_f) + H_i HML + S_i SMB + \epsilon_i
$$

其中HML是高B/M比例者（H）减低者(L)的“超额收益率”，SMB是小(S)减大者(B)的“超额收益率”。

我们可以取最近两三年时间，取每月的收益率（同时考虑资本利得与股利）$r_i$，与每年定期存款利率的月化版本为无风险利率$r_f$，每月的大盘指数，如上证指数收益率为市场组合收益率的代理$r_m$，HML,SMB可以寻求其他更专业的变量，也可以光依靠财务数据对其进行排序分类，然后计算其平均值作为代理。这三个变量所形成的时间序列可用于所有个股，每个个股再带一个自己的收益率时间序列就可以估计出自己的三个beta参数与alpha了。我们对那些alpha值为正的股票特别感兴趣，这也被称为“寻找alpha”（超额收益率）的行动，虽然这是个过于简单的模型，但它的确找出了一个alpha。

对于从众多股票中选择10只股票的任务，我选择寻找alpha最大以及考虑标准差后alpha与标准差之比最大化的那些股票。虽然这并不是很正确的方法，显然市场组合包含的股票越多越好，这样才能尽可能抵消非系统风险，限制为只能选取10只未免破坏了抵消非系统风险和全局优化的机会。
